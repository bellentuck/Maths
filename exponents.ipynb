```python
import matplotlib.pyplot as plt
```

```python
bases = [i for i in range (0, 101)]
exponents = [i for i in range(0, 11)]
```

```python
for exp in exponents:
    plt.plot([b for b in bases], [b**exp for b in bases], '--', linewidth=2)
    plt.show()
```
